```python
from secret import token
from lyricsgenius import Genius
import pandas as pd
```

```python
genius = Genius(token, skip_non_songs=True, excluded_terms=["(Remix)", "(Live)", "(Radio)", "(Demo)"], remove_section_headers=True)
```

```python
def get_lyrics(artist, num_songs):
    songs = genius.search_artist(artist, max_songs=num_songs, sort='popularity').songs

    # put the data in a dict
    data = {"title": [], "lyrics": []}
    for song in songs:
        data["title"].append(song.title)
        data["lyrics"].append(song.lyrics)

    # save lyrics to csv
    df = pd.DataFrame.from_dict(data)
    file_name = "lyrics/" + artist.replace(" ", "_") + ".csv"
    df.to_csv(file_name, index=False)

    print(f"{df.shape[0]} {artist} songs saved to {file_name}")
    
```

```python
bob_dylan = get_lyrics("Bob Dylan", 500)
```

```python
taylor_swift = get_lyrics("Taylor Swift", 300)
```
